```python
import os
import glob
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
from matplotlib import rc
plt.rcParams['font.family'] = 'AppleGothic'
rc('axes', unicode_minus=False)
```

```python
train_input_csv_paths = sorted(glob.glob('./data/train_input/*.csv'))
train_target_csv_paths = sorted(glob.glob('./data/train_target/*.csv'))

test_input_csv_paths = sorted(glob.glob('./data/test_input/*.csv'))
```

```python
train_input_df_list = []
for path in train_input_csv_paths:
    print(">>", path)
    input_df = pd.read_csv(path)
    train_input_df_list.append(input_df)

    print(input_df.describe())
    print(input_df.isnull().sum())
    print(input_df.loc[pd.isna(input_df).any(1), :].index)
```

```python
for df in train_target_df_list:
    df.plot()
```

```python
index = 0
#.drop(['내부습도관측치', '일간누적분무량', '냉방부하', '레드 LED상태'], axis =1)
input_df = train_input_df_list[index]
target_df = train_target_df_list[index]

ax = input_df.plot()
target_df.plot(ax = ax)

plt.rcParams["figure.figsize"] = [20, 10]
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
